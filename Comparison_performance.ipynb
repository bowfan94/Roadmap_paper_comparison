```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
```

```python
performance_all = pd.read_csv('./performance.csv',index_col=0)
performance_all
```

```python
x = [0, 1]
plt.figure(figsize=(10, 16))
red_count = 0
green_count = 0
for i in range(len(performance_all)):
    c = 'r' if performance_all.iloc[i, 3] >= performance_all.iloc[i, 0] else 'g'
    plt.plot(x, performance_all.iloc[i, [0, 3]], color=c)

plt.title('PRS-MRS', fontsize=30)
plt.ylabel('AUROC', fontsize=20)
plt.xticks([0,1],['PRS', 'MRS'],fontsize=20)
plt.tight_layout()
plt.show()
```
